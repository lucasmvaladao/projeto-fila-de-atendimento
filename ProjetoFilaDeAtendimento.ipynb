```python
filaNormal = []
filaPreferencial = []
historico = {}
contadorN = 1
contadorP = 1
while True:
  print("""Bem-vindo ao sistema de atendimento bancário!
  1 - Adicionar cliente
  2 - Atender cliente
  3 - Exibir filas
  4 - Exibir histórico de atendimentos
  5 - Sair""")

  opcao = input("\nEscolha uma opção: ")

  if opcao == "1":
      nome = str(input("Digite o nome do cliente: "))

      while True:
        status = str(input("O cliente é preferencial? (S/N): "))
        if status.upper() == "S" :
          filaPreferencial.append(nome)

          print(f"\nCliente {nome} adicionado com sucesso a fila preferencial!\n")
          break
        elif status.upper() == "N":
          filaNormal.append(nome)
          print(f"\nCliente {nome} adicionado com sucesso a fila normal!\n")
          break
        else:
           print("Opção inválida. Por favor, digite apenas 'S' para sim ou 'N' para não.\n")


  elif opcao == "2":
        print("Atendendo cliente...\n")

        if filaPreferencial and contadorP > 0:
          removido = filaPreferencial.pop(0)
          print(f"{removido} chegou sua vez. \n")
          contadorP -= 1
          historico[removido] = "Preferencial"
        elif filaNormal:
            removido = filaNormal.pop(0)
            print(f"{removido} chegou sua vez. \n")
            contadorN -= 1
            historico[removido] = "Normal"

        if contadorN == 0 and contadorP == 0:
          contadorN = 1
          contadorP = 1

  elif opcao == "3":
      print("Exibindo filas...\n")

      if not filaPreferencial:
        print(f"A fila preferencial está vazia! \n")
      else:
        print("Situação da fila preferencial: ")
        print(filaPreferencial)
        print("")

      if not filaNormal:
        print(f"A fila normal está vazia! \n")
      else:
        print("Situação da fila normal: ")
        print(filaNormal)
        print("")

  elif opcao == "4":
      print("Exibindo histórico de atendimentos... \n")
      for nome, status in historico.items():
        print(f"Nome: {nome}")
        print(f"Status: {status}")
        print("-" * 10)

  elif opcao == "5":
      print("Saindo do sistema. Até logo!")
      break

  else:
      print("Opção inválida!")
```
